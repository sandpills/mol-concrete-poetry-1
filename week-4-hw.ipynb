Import libraries, set up HTML template

```python
import math
import random
import numpy as np

from IPython.display import display, HTML
def show_html(src):
    return display(HTML(src), metadata=dict(isolated=True))

html_tmpl = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="utf-8">
    <title>{title}</title>
    <style>
    html {{ overflow: visible; }}
    </style>
</head>
<body>
{content}
</body>
</html>"""
```

```python
def mkdiv(content, **kwargs):
    if 'position' not in kwargs:
        kwargs['position'] = 'absolute'
    style_str = ' '.join([": ".join((k.replace('_', '-'), v))+";" for k, v in kwargs.items()])
    return f"<div style='{style_str}'>{content}</div>"
```

Below are my attempts to create a composition with "you" "I" and "water" based on the top-spin examples

```python
text = "我  水  你"
s = text.replace('水', '水'*12, 1)

print (s)
```

```python
def sinerange(n, steps=10):
    for i in range(steps):
        theta = (math.tau / steps) * i
        yield int(n/2 * (-1*math.cos(theta))) + int(n/2)

def sinezigzag(s, chunklen=2, steps=20):
    out = []
    for i in sinerange(len(s)-(chunklen-1), steps):
        out.append(" " * i + s[i:i+chunklen])
    return "\n".join(out)

for i in range(1):
    content = sinezigzag(s, 4, 10)  
    print (content)
```

# “Tree” 木

A forrest comprised of human (人), wood (木) and ten (十)

I'd like to add some variety in tree positions but don't really know how since this is a tripple for loop...

```python
text = "人木十"
wood = text.replace('木', '木'*3, 1)

grid_size_x = 20
grid_size_y = len(wood)

divs = []
for n in range(5):
    for i in range(grid_size_y):
        for j in range(grid_size_x):
            x_pos = 7 * i
            y_pos = (90 / grid_size_x) * j
            this_div = mkdiv(wood[i],
                             top=f"{x_pos}",
                             left=f"{y_pos}%",
                             height="1em")
            divs.append(this_div)
    html_src = html_tmpl.format(title="frost grid", content="".join(divs))
    show_html(html_src)
```

There are so many idions with four characters that start with a "one", which looks like a little stick. Sans Serif fonts make them look straight and flat, while Serif fonts look spiky and impulsive. Below are my attempts of arranging them in different circular shapes.

After I got rid of the "minimum-height" property in the html template they stopped being properly displayed :(

```python
idioms = [
  "一见钟情",
  "一表人才",
  "一蹴而就",
  "一马当先",
  "一无所有",
  "一事无成",
  "一技之长",
  "一石二鸟",
  "一举两得",
  "一步之遥",
  "一骑绝尘",
  "一望无垠",
  "一五一十",
  "一视同仁",
  "一眼万年",
  "一呼百应",
  "一刀两断",
  "一本正经"
]
divs = []
a = len(idioms)
for i in range(a):
    degrees = i * (360 / a) 
    this_div = mkdiv(idioms[i],
                      top="50%",
                      left="20%",
                      width="1em",
                      height= "15em",
                      font_family="Helvetica",
                      transform=f"translate(-50%, -50%) rotate({degrees}deg)")
    divs.append(this_div)
html_src = html_tmpl.format(title="centered", content="".join(divs))
show_html(html_src)
```
